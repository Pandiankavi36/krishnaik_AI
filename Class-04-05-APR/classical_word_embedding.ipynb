```python
from sklearn.preprocessing import OneHotEncoder
import numpy as np
```

```python
documents =["people watch movie", 
            "people watch cricket", 
            "people like movie",
            "people like cricket"
            ]
```

```python
tokens = [sentence.lower().split() for sentence in documents]
print(tokens)
```

```python
tokens
```

conver the tokens into 2D word matrix

```python
all_words = [[word] for sentence in tokens for word in sentence]
```

```python
all_words
```

creating object for OHE

OneHotEncoder is a predefined class 

```python
encoder = OneHotEncoder(sparse_output=False)
```

```python
encoder.fit(all_words)
```

```python
encoder.categories_
```

```python
print("Vocabulary:",encoder.categories_[0])
```

```python
tokens
```

onvert tokens into vectors using encoder

```python
for sentence in tokens:
    encoded_sentence = encoder.transform([[word] for word in sentence])
    print("\nSentence:", sentence)
    print( encoded_sentence)
```

BOW - Bag of Worrds 

Convert a collection of text documents to a matrix of token counts.

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python

documents =["people watch movie and watch movie again",
"people watch cricket and watch cricket",
"people like movie and like movie a lot",
"people like cricket"]
```

```python
documents
```

AIM: convert data into NUmber

```python
bow = CountVectorizer()
```

```python
bow
```

```python
bow_matrix_fit = bow.fit(documents)
```

```python
bow_matrix_fit
```

```python
bow_matrix = bow.fit_transform(documents)
```

```python
bow_matrix
```

```python
print("BOW Matrix:\n",bow_matrix.toarray())
```

to get the vocabulary

```python
bow.get_feature_names_out()
```

```python
print("vocabulary:", bow.get_feature_names_out())
```
