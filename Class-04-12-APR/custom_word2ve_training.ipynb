```python
folder_path = current_dir/"data"
story = []
for filename in os.listdir(folder_path):
    file_path = os.path.join(folder_path, filename)
    with open(file_path, 'r', encoding ='unicode_escape') as f:
        corpus = f.read()
    raw_sent = sent_tokenize(corpus)
    for sent in raw_sent:
        story.append(simple_preprocess(sent))    
   
```

```python
custom_models = gensim.models.Word2Vec(story, vector_size=150, window=10, min_count=1)
```

```python
custom_models
```

```python
custom_models.build_vocab(story)
```

```python
custom_models.corpus_count
```

```python
custom_models.train(story, total_examples=custom_models.corpus_count, epochs=10)
```
